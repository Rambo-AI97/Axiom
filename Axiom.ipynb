```python
import os
import json
import requests

class Axiom:
    def __init__(self, root_dir=None):
        self.root_dir = root_dir or os.getcwd()
        self.soul_path = os.path.join(self.root_dir, "soul.txt")
        self.ltm_path = os.path.join(self.root_dir, "long_term_memory.json")
        self._boot_sequence()

    def _boot_sequence(self):
        if not os.path.exists(self.soul_path):
            with open(self.soul_path, "w") as f:
                f.write("You are Axiom, a Data Scientist. Goal: Self-improve code.")
        if not os.path.exists(self.ltm_path):
            with open(self.ltm_path, "w") as f:
                json.dump([], f)

    def get_available_models(self):
        try:
            res = requests.get('http://localhost:11434/api/tags')
            return [m['name'] for m in res.json().get('models', [])]
        except:
            return ["mistral"] # Fallback

    def read_files(self):
        knowledge = {}
        ignore = ['response.json', 'long_term_memory.json', '.git', '.venv']
        
        for file in os.listdir(self.root_dir):
            if any(file.endswith(ext) for ext in ['.pyc', '.bin']) or file in ignore:
                continue
            
            path = os.path.join(self.root_dir, file)
            if os.path.isfile(path):
                try:
                    with open(path, "r", encoding="utf-8") as f:
                        content = f.read()
                        # Better Notebook handling: Extract code cells only
                        if file.endswith('.ipynb'):
                            nb = json.loads(content)
                            content = "\n".join(["".join(c['source']) for c in nb['cells'] if c['cell_type'] == 'code'])
                        knowledge[file] = content[:3000]
                except:
                    continue
        return knowledge

    def call_llm(self, model, system_prompt, user_prompt):
        url = 'http://localhost:11434/api/generate'
        payload = {
            "model": model,
            "prompt": f"SYSTEM: {system_prompt}\n\nUSER: {user_prompt}",
            "format": "json",
            "stream": False
        }
        response = requests.post(url, json=payload, timeout=300)
        return json.loads(response.json().get('response', '{}'))
```
